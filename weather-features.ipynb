```python
rf_df = df.drop(['NAME', 'SNOW_ATTRIBUTES', 'SNWD_ATTRIBUTES', 'TAVG_ATTRIBUTES', 'TMAX_ATTRIBUTES', 'TMIN_ATTRIBUTES', 'SNOW', 'SNWD', 'LATITUDE', 'LONGITUDE', 'ELEVATION'], axis=1)
```

```python
for station in rf_df['STATION'].unique():
    print(station)
    print(rf_df[rf_df['STATION'] == station].describe())
```

```python
# do a KNN imputation on the missing values
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)

# only impute the missing values in the TAVG, TMAX, TMIX column
imputer.fit(rf_df[['TAVG', 'TMAX', 'TMIN']])
imputed_df = imputer.transform(rf_df[['TAVG', 'TMAX', 'TMIN']])
rf_df[['TAVG', 'TMAX', 'TMIN']] = imputed_df
rf_df.plot(x='DATE', y=['TAVG', 'TMAX', 'TMIN'], figsize=(15, 5))
```

```python
# upload the data to the database

import psycopg2.extras
from db import DBManager
from pprint import pprint

values = rf_df.values.tolist()
pprint(values)
print(len(values))  

with DBManager() as db:
    conn = db.conn
    cur = conn.cursor()
    cur.execute('DROP TABLE IF EXISTS weather')
    cur.execute("""
        CREATE TABLE weather (
            station varchar(255),
            date date,
            tavg float,
            tmax float,
            tmin float
        )
    """)
    psycopg2.extras.execute_values(cur, """
        INSERT INTO weather (station, date, tavg, tmax, tmin) VALUES %s
    """, values)
    conn.commit()
    
```
